```python
import turicreate as tc
import numpy as np
```

```python
sales = tc.SFrame('./kc_house_data.gl/')
sales
```

```python
train_data, test_data = sales.random_split(.8,seed=0)
input_feature = train_data['sqft_living']
output = train_data['price']
```

```python
def simple_linear_regression(input_feature, output):
    # quickly assign x and y
    x, y = input_feature.to_numpy(), output.to_numpy()
#     x, y = input_feature, output

    # size of dataset 
    n = np.size(x)
    
    # calculate sums
    sum_x, sum_y = np.sum(x), np.sum(y)
    sum_xx, sum_xy  = np.sum(x**2), np.sum(x*y)
    
    num = (sum_xy) - (1/n)*(sum_x*sum_y)
    denom = (sum_xx) - (1/n)*(sum_x*sum_x)
    
    # calculating regression coefficients
    b_1 = num / denom
    b_0 = np.mean(y) - b_1*np.mean(x)
    
    return b_0,b_1
    
squarefeet_intercept, squarefeet_slope = simple_linear_regression(input_feature, output)
# squarefeet_intercept, squarefeet_slope = simple_linear_regression(np.array([0,1,2,3,4]), np.array([1,3,7,13,21]))

print(squarefeet_intercept, squarefeet_slope)
```

```python
def get_regression_predictions(input_feature, intercept, slope):
    # list comprehension, predictions for each value using given intercept and slope
    return [slope*value+intercept for value in input_feature]

# predicted_output = get_regression_predictions(input_feature,squarefeet_intercept,squarefeet_slope)
# print(predicted_output)
```

```python
answer_1 = get_regression_predictions([2650], squarefeet_intercept, squarefeet_slope)
print(round(answer_1[0],2))
```

```python
def get_residual_sum_of_squares(input_feature, output, intercept, slope):
    # get predictions
    predictions = get_regression_predictions(input_feature, intercept, slope)
    y = output.to_numpy()
    
    # calculate residual
    residual = y - predictions
    
    return np.sum(residual**2)

answer_2 = get_residual_sum_of_squares(input_feature, output, squarefeet_intercept, squarefeet_slope)
```

```python
print(answer_2)
```

```python
def inverse_regression_predictions(output, intercept, slope):
    return (output-intercept)/slope

# inv = inverse_regression_predictions(output,squarefeet_intercept,squarefeet_slope)
# print(inv)
```

```python
input_feature = train_data['bedrooms']
bedroom_intercept, bedroom_slope = simple_linear_regression(input_feature, output)

print(bedroom_intercept, bedroom_slope)
```
